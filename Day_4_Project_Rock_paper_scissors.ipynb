```
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

import random
player_input = int(input ('What do you choose? Type 0 for Rock, 1 for Paper, 2 for Scissors '))

if player_input < 0 or player_input >= 3:
  print ('You chose an invalid number, You lost!')
elif player_input == 0:
  print ('You chose:')
  print (rock)
elif player_input == 1:
  print ('You chose:')
  print (paper)
elif player_input == 2:
  print ('You chose:')
  print(scissors)

comp_input = random.randint(0,2)

if comp_input == 0 and player_input == 0:
  print ('The computer chose:')
  print (rock)
  print ('Draw!')
elif comp_input == 0 and player_input == 1:
  print ('The computer chose:')
  print (rock)
  print ('You Won!')
elif comp_input == 0 and player_input == 2:
  print ('The computer chose:')
  print (rock)
  print ('You Lost!')

if comp_input == 1 and player_input == 0:
  print ('The computer chose:')
  print (paper)
  print ('You Lost!')
elif comp_input == 1 and player_input == 1:
  print ('The computer chose:')
  print (paper)
  print ('Draw!')
elif comp_input == 1 and player_input == 2:
  print ('The computer chose:')
  print (paper)
  print ('You Won!')

if comp_input == 2 and player_input == 0:
  print ('The computer chose:')
  print (scissors)
  print ('You Won!')
elif comp_input == 2 and player_input == 1:
  print ('The computer chose:')
  print (scissors)
  print ('You Lost!')
elif comp_input == 2 and player_input == 2:
  print ('The computer chose:')
  print (scissors)
  print ('Draw!')
```
